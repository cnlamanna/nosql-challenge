### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}
Score_of_20 = establishments.find(query)
# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print("number of establishments w/20 score:", count)
# Display the first document in the results using pprint
for establishment in Score_of_20:
    pprint(establishment)
```

```python
query = {"scores.Hygiene": 20}
Score_of_20 = establishments.find(query)
# Convert the result to a Pandas DataFrame
score_of_20_list = list(Score_of_20)
df = pd.DataFrame(score_of_20_list)
# Display the number of rows in the DataFrame
num_rows = len(df)
print("number of rows:", num_rows)
# Display the first 10 rows of the DataFrame
first_10_rows = df.head(10)
print(first_10_rows)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
PF_latitude = 51.49014200
PF_longitude = 0.08384000

query2 = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": PF_latitude - degree_search, "$lte": PF_latitude + degree_search},
    "geocode.longitude": {"$gte": PF_longitude - degree_search, "$lte": PF_longitude + degree_search}
}
 
sort =  [("scores.Hygiene",1)]

# Print the results
top_5 = establishments.find(query2).sort(sort).limit(5)


for establishment2 in top_5:
    pprint(establishment2)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {"count": -1}
    }
]
result4 = list(establishments.aggregate(pipeline))
# Print the number of documents in the result
print("number of docs:", len(result4))
# Print the first 10 results
for item in result4[:10]:
    pprint(item)
```

```python

# Convert the result to a Pandas DataFrame
Last_Results = list(result4)
Last_Results_df = pd.DataFrame(Last_Results)
# Display the number of rows in the DataFrame
num_rows4 = len(Last_Results)
print("number of rows:", num_rows4)
# Display the first 10 rows of the DataFrame
Last_10 = Last_Results_df.head(10)
print(Last_10)
```
